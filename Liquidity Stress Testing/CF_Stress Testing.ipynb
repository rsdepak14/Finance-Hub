```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
days = 90
initial_cash = 5_000_000          # $5 million starting balance
threshold = 1_000_000             # $1 million minimum liquidity threshold

# Daily cash flow assumptions
normal_inflow = 500_000           # average daily receivable (in USD)
normal_outflow = 480_000          # average daily payment
stress_inflow = 400_000           # stressed daily receivable (reduced due to delays/FX)
stress_outflow = 550_000          # stressed daily payment (increased/accelerated)

# Simulate random daily flows (10% standard deviation for variability)
np.random.seed(0)  # for reproducibility
normal_inflows = np.random.normal(normal_inflow, 0.1*normal_inflow, days)
normal_outflows = np.random.normal(normal_outflow, 0.1*normal_outflow, days)
stress_inflows = np.random.normal(stress_inflow, 0.1*stress_inflow, days)
stress_outflows = np.random.normal(stress_outflow, 0.1*stress_outflow, days)

# Calculate daily cash balances for each scenario
normal_balance = initial_cash + np.cumsum(normal_inflows - normal_outflows)
stress_balance = initial_cash + np.cumsum(stress_inflows - stress_outflows)

# Plot the results
plt.figure(figsize=(6,4))
plt.plot(normal_balance, label='Normal')
plt.plot(stress_balance, label='Stressed')
plt.axhline(threshold, color='r', linestyle='--', label='Threshold')
plt.xlabel('Day'); plt.ylabel('Cash Balance (USD)')
plt.legend(); plt.tight_layout()
plt.show()

```
