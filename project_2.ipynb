## Executive Summary


### Objective
This project provides a detailed evaluation of three trading strategies applied to the Dow Jones Industrial Average ETF (DIA) and its underlying components over a 27-year period, from January 20, 1998, to December 31, 2024. The goal is to critically compare passive investing with active, breakout-based strategies to understand their performance across different market conditions.

The analysis examines how simple, rule-based breakout entry and exit signals stack up against a traditional buy-and-hold approach. Rather than focusing solely on absolute returns, it also considers risk-adjusted performance, drawdowns, and long-term portfolio growth. Equal weight is given to assessing how practical, resilient, and adaptable each strategy is during key periods like the dot-com bubble, the 2008 financial crisis, and the post-COVID market recovery.

### Comparison and Results Interpretation
#### A. Performance Metrics [Table 1]
The passive Buy and Hold DIA strategy delivered a Total Return of 855.93% and a CAGR of 8.74%. Its Sharpe (0.44) and Sortino (0.56) ratios reflect risk-adjusted returns and relatively controlled downside risk. Strategy 1 has the market beta of 1. Strategy 2 has a negative beta (-0.16), meaning its returns move slightly opposite to the market. This reflects its defensive design—it avoids major drops by exiting during periods of high risk, helping to reduce drawdowns and preserve capital. Strategy 3 also shows a negative beta (-0.19), suggesting a low and slightly inverse correlation with the broader market. However, its exposure to individual DJIA stocks introduces higher volatility, which explains its deeper drawdowns during extreme events.

The Breakout Strategy on DIA ETF performed poorly, with a Total Return of -16.86% and CAGR of -0.68%. The negative Sharpe and Sortino ratios suggest the strategy suffered from poor timing and too many false signals. Strategy 3, which applied breakout signals to individual DIA stocks, did much better than Strategy 2. It earned a Total Return of 367.25% and a CAGR of 5.89%, showing that trading individual stocks can be more effective than the ETF. Still, its Sharpe and Sortino  ratios were lower than the passive strategy, meaning higher volatility and less efficient returns.

#### B. Comparison of Strategies in Crises Periods [Figure 6]
Shaded areas show major financial crises: the 2000 Dot-Com Bubble, 2008 Financial Crisis, and 2020 COVID Crash. Crises and Recovery: 2000 & 2008: Strategies 1 and 3 fall ~50% but recover in 4–5 years. Strategy 2: Drops only ~20% and stays flat — it avoids big losses but also misses rebounds. 2020 Crash: All strategies take a hit, but 1 and 3 bounce back fast. Strategy 2 stays mostly flat.

#### C. Drawdown Comparison: Buy & Hold vs Breakout Strategies [Figure 7 and Table 2]
Strategy 1 (Buy & Hold DIA) aligns with the Efficient Market Hypothesis (EMH), providing full market exposure but suffering from deep and prolonged drawdowns, such as -33.95% in 2000 (771-day recovery) and -51.87% in 2008. In contrast, Strategy 2 (Breakout on DIA) and Strategy 3 (Breakout on DJIA Constituents) attempt to exploit market inefficiencies. Strategy 2 shows smaller drawdowns (-35.03% in 2000, -34.84% in 2008) than Strategy 3, while Strategy 3, despite managing risk dynamically across stocks, still experiences large swings (-49.47% in 2000, -36.48% in 2008). Overall, the breakout strategies aim to reduce exposure during downturns and capitalize on recoveries, making them more adaptive.

### Conclusion
Passive investing, such as a buy-and-hold approach in the DJIA ETF, often outperforms active strategies due to reduced biases and the ability to capture long-term market growth. The breakout strategy, despite its intent to capitalize on short-term price movements, faces risks like false breakouts, higher volatility, and missed opportunities during sustained bull markets. 

Additionally, active strategies require continuous monitoring and adjustments, which may not always justify the additional effort when broad market exposure historically delivers consistent returns. Over time, the compounding effect of staying invested in the market outweighs the potential gains from frequent trading, making passive investing a more reliable wealth-building approach.

```python
## Breakout Strategy Function [to utilize in Strategies 2 & 3]
```

```python
def simulate_breakout_trading(stock_inputdata, initial_cash):
    """
    Simulates the 52-week high/low breakout strategy for a single stock.

    Parameters:
    stock_inputdata (DataFrame): Data for a single stock.
    initial_cash (float): Starting capital.

    Returns:
    DataFrame: Contains strategy performance metrics.
    """

    # Calculate technical indicators
    stock_data = stock_inputdata.copy()
    stock_data['52_week_high'] = stock_data['Adj Close'].rolling(window=252).max()
    stock_data['52_week_low'] = stock_data['Adj Close'].rolling(window=252).min()
    stock_data['20_day_SMA_vol'] = stock_data['Volume'].rolling(window=20).mean()

    # Define entry signals
    stock_data['Long_Entry'] = (stock_data['Adj Close'] >= stock_data['52_week_high']) & (stock_data['Volume'] > 1.5 * stock_data['20_day_SMA_vol'])
    stock_data['Short_Entry'] = (stock_data['Adj Close'] <= stock_data['52_week_low']) & (stock_data['Volume'] > 1.5 * stock_data['20_day_SMA_vol'])

    # Initialize tracking variables
    stock_data['Position'] = 0  

    position = 0
    entry_price = 0
    entry_idx = None

    # Iterate through the DataFrame
    for i in range(1, len(stock_data)):
        current_price = stock_data.iloc[i]['Adj Close']

        if position == 0:
            # Long Entry
            if stock_data.iloc[i]['Long_Entry']:
                position = 1
                entry_price = current_price
                entry_idx = i

            # Short Entry
            elif stock_data.iloc[i]['Short_Entry']:
                position = -1
                entry_price = current_price
                entry_idx = i

        elif position == 1:
            # Check exit conditions for long
            if current_price < 0.95 * entry_price or (i - entry_idx) >= 20:
                # Before exiting, check if Long_Entry is true again
                if stock_data.iloc[i]['Long_Entry']:
                    position = 1  # Re-enter immediately
                    entry_price = current_price
                    entry_idx = i
                else:
                    position = 0
                    entry_idx = None

        elif position == -1:
            # Check exit conditions for short
            if current_price > 1.05 * entry_price or (i - entry_idx) >= 20:
                # Before exiting, check if Short_Entry is true again
                if stock_data.iloc[i]['Short_Entry']:
                    position = -1  # Re-enter immediately
                    entry_price = current_price
                    entry_idx = i
                else:
                    position = 0
                    entry_idx = None

        stock_data.at[stock_data.index[i], 'Position'] = position  # Ensures position is correctly carried forward

    # Calculate strategy returns
    stock_data['Daily_Return'] = stock_data['Adj Close'].pct_change()

    return stock_data
```

```python
## Breakout Strategy - DJIA ETF [Strategy 2]
```

```python
# Run the breakout strategy simulation for DIA
stock_inputdata = djia.loc['DIA']
Strat2 = simulate_breakout_trading(stock_inputdata, initial_cash)

# Compute strategy and cash returns
Strat2['Strategy_Return'] = Strat2['Daily_Return'] * Strat2['Position'].shift(1)
Strat2['Cash_Return'] = Strat2['RF'] / 100
Strat2['Total_Return'] = Strat2['Strategy_Return'].fillna(0) + Strat2['Cash_Return'] * (Strat2['Position'].shift(1) == 0)

# Fix first row separately to avoid SettingWithCopyWarning
Strat2.loc[Strat2.index[0], 'Total_Return'] = Strat2.loc[Strat2.index[0], 'RF'] / 100

# Initialize cumulative returns starting at 10000
Strat2['Cumulative_Return'] = (Strat2['Total_Return'] + 1).cumprod() * 10000

# Ensure the first day's value is exactly 10000 using .loc[]
Strat2.loc[Strat2.index[0], 'Cumulative_Return'] = 10000*(1+(Strat2['RF'].iloc[0])/100)
```

```python
# Calculate and print performance metrics
total_return = (1 + Strat2['Total_Return']).prod() - 1  # Total return in decimal

# Compute daily returns
daily_returns = Strat2['Total_Return'].dropna()  

# Sharpe Ratio (Annualized)
sharpe = (daily_returns.mean() - rf.mean()) / daily_returns.std() * (252 ** 0.5)

# # Print metrics
# print("Breakout Strategy (DIA) Performance Metrics:")
# print(f"Total Return: {total_return*100:.2f}%")
# print(f"Sharpe Ratio: {sharpe:.2f}")
```

```python
## Breakout Strategy - Individual Constituents [Strategy 3]
```

```python
# Strategy 3 - Breakout Strategy on DJIA Constituents
# Apply the breakout strategy to all tickers (excluding DIA)
all_stocks = djia.drop(index='DIA', level='Ticker')
all_stocks = all_stocks.groupby('Ticker').apply(simulate_breakout_trading, initial_cash)

# Get all unique dates in the dataset and sort them
unique_dates = sorted(all_stocks.index.get_level_values('Date').unique())

# Initialize results list to store daily returns
strat3_results = []
```

```python
# First day: no prior positions, so earn risk-free rate
first_date = unique_dates[0]
first_rf = all_stocks.xs(first_date, level='Date')['RF'].iloc[0] / 100
strat3_results.append([first_date, first_rf])
```

```python
# Iterate over all dates, starting from the second date
for i, date in enumerate(unique_dates[1:]):
    # Find the previous date (the one before the current date)
    prev_date = unique_dates[i]

    # Select data for the current date and the previous date
    target_loc = all_stocks.xs(date, level='Date')
    prev_target_loc = all_stocks.xs(prev_date, level='Date')

    # Counting active positions (non-zero)
    active_mask = (prev_target_loc['Position'] != 0)
    num_active_positions = active_mask.sum()

    # Check if there are any active positions
    if num_active_positions == 0:
        cash_return = target_loc['RF'] / 100
        total_return = cash_return.iloc[0]  # Since RF is the same for all stocks on a given date
    else:
        # Calculate the average return of active stocks
        active_positions = prev_target_loc['Position'][active_mask]
        active_returns = target_loc['Daily_Return'][active_mask]
        total_return = np.sum(active_positions * active_returns) / num_active_positions

    # Store results in list (date, return)
    strat3_results.append([date, total_return])
```

```python
# Convert results to a DataFrame
Strat3 = pd.DataFrame(strat3_results, columns=['Date', 'Total_Return'])
Strat3.set_index('Date', inplace=True)

# Compute cumulative returns starting at initial_cash
Strat3['Cumulative_Return'] = (1 + Strat3['Total_Return']).cumprod() * initial_cash

# Calculate performance metrics
strat3_total_return = (Strat3['Cumulative_Return'].iloc[-1] - initial_cash) / initial_cash * 100
strat3_sharpe = (Strat3['Total_Return'].mean() - (all_stocks['RF'] / 100).mean()) / Strat3['Total_Return'].std() * (252 ** 0.5)

# # Print metrics
# print("Breakout Strategy (DJIA Constituents) Performance Metrics:")
# print(f"Initial Cash: ${initial_cash:.2f}")
# print(f"Final Portfolio Value: ${Strat3['Cumulative_Return'].iloc[-1]:.2f}")
# print(f"Total Return: {strat3_total_return:.2f}%")
# print(f"Sharpe Ratio: {strat3_sharpe:.2f}")
```

### Performance Metrics

```python
def calculate_strategy_metrics(strategy_dfs: dict):
    results = {}

    rf_series = strategy_dfs['Strat1']['RF'] / 100

    for strat_name, df in strategy_dfs.items():
        df = df.dropna(subset=["Total_Return", "Cumulative_Return"]).copy()

        daily_return = df["Total_Return"]
        excess_return = daily_return - rf_series[:len(daily_return)].values
        total_return = ((df["Cumulative_Return"].iloc[-1] / df["Cumulative_Return"].iloc[0]) - 1) * 100
        num_years = (df.index[-1] - df.index[0]).days / 365.25
        cagr = ((df["Cumulative_Return"].iloc[-1] / df["Cumulative_Return"].iloc[0]) ** (1 / num_years) - 1) * 100

        # Beta
        beta = np.cov(df['Total_Return'], strategy_dfs['Strat1']['Total_Return'])[0, 1] / np.var(strategy_dfs['Strat1']['Total_Return'])
        # Sharpe Ratio
        sharpe_ratio = (np.mean(excess_return) / np.std(daily_return)) * np.sqrt(252)
        # Treynor Ratio
        treynor_ratio = (np.mean(excess_return) * 252) / beta if beta != 0 else np.nan
        # Information Ratio
        market_return = strategy_dfs['Strat1']["Total_Return"].dropna()[:len(daily_return)]
        active_return = daily_return - market_return
        tracking_error = np.std(active_return)
        info_ratio = (np.mean(active_return) / tracking_error) * np.sqrt(252) if tracking_error != 0 else np.nan
        # Sortino Ratio
        downside = daily_return[daily_return < 0]
        downside_std = np.std(downside)
        sortino_ratio = (np.mean(excess_return) / downside_std) * np.sqrt(252) if downside_std != 0 else np.nan

        # Store results rounded
        results[strat_name] = {
            "Total Return (%)": total_return,
            "CAGR (%)": cagr,
            "Beta": beta,
            "Sharpe Ratio": sharpe_ratio,
            "Treynor Ratio": treynor_ratio,
            "Information Ratio": info_ratio,
            "Sortino Ratio": sortino_ratio
        }

    # Create DataFrame and style it
    df_metrics = pd.DataFrame(results).T.round(2)

    styled_table = df_metrics.T.style\
        .format("{:.2f}")\
        .set_caption("Strategy Performance Metrics")\
        .set_properties(**{
            'text-align': 'center',
            'font-family': 'Arial',
            'font-size': '13px'
        })\
        .set_table_styles([
            {'selector': 'caption', 'props': [
                ('color', '#000'),
                ('font-size', '15px'),
                ('text-align', 'left'),
                ('font-weight', 'bold'),
                ('padding', '10px 0')
            ]},
            {'selector': 'th', 'props': [('text-align', 'center'), ('border', '1px solid black')]},
            {'selector': 'td', 'props': [('border', '1px solid black')]},
            {'selector': 'table', 'props': [('border-collapse', 'collapse')]}
        ])

    return styled_table
    
strategy_dfs = {
    'Strat1': Strat1,
    'Strat2': Strat2,
    'Strat3': Strat3
}
```

### Crisis Analysis

```python
# Function to compute metrics for a given period
def compute_crisis_metrics(Cumulative_Return, start_date, end_date, daily_returns):
    # Filter data for the crisis period
    period_data = Cumulative_Return[(Cumulative_Return.index >= start_date) & (Cumulative_Return.index <= end_date)]
    if period_data.empty:
        return None, None, None, None

    # Drawdown: Maximum percentage loss from peak to trough
    running_max = period_data.cummax()
    drawdowns = (period_data - running_max) / running_max
    max_drawdown = drawdowns.min() * 100  # Convert to percentage

    # Recovery Time: Days to recover to pre-crisis peak (or None if not recovered)
    pre_crisis_peak = Cumulative_Return[Cumulative_Return.index < start_date].iloc[-1] if not Cumulative_Return[Cumulative_Return.index < start_date].empty else initial_cash
    post_crisis_data = Cumulative_Return[Cumulative_Return.index >= end_date]
    recovery_date = post_crisis_data[post_crisis_data >= pre_crisis_peak].index
    recovery_time = (recovery_date[0] - pd.to_datetime(end_date)).days if not recovery_date.empty else None

    # Total Return: Percentage change over the period
    start_value = period_data.iloc[0]
    end_value = period_data.iloc[-1]
    total_return = (end_value / start_value - 1) * 100

    return max_drawdown, recovery_time, total_return
```

```python
# Compute daily returns for each strategy
strat1_daily_returns = Strat1['Cumulative_Return'].pct_change().dropna()
strat2_daily_returns = Strat2['Total_Return'].dropna()  # Already computed in Strategy 2
strat3_daily_returns = Strat3['Total_Return'].dropna()

# Compute metrics for each crisis period and strategy
metrics_data = []
for crisis, info in crisis_periods.items():
    start_date, end_date = info["dates"]
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)

    # Strategy 1
    strat1_metrics = compute_crisis_metrics(Strat1['Cumulative_Return'], start_date, end_date, strat1_daily_returns)
    strat2_metrics = compute_crisis_metrics(Strat2['Cumulative_Return'], start_date, end_date, strat2_daily_returns)
    strat3_metrics = compute_crisis_metrics(Strat3['Cumulative_Return'], start_date, end_date, strat3_daily_returns)

    metrics_data.append({
        'Crisis Period': crisis,
        'Strategy': 'Buy & Hold (DIA)',
        'Max Drawdown (%)': strat1_metrics[0],
        'Recovery Time (Days)': strat1_metrics[1],
        'Total Return (%)': strat1_metrics[2]
    })
    metrics_data.append({
        'Crisis Period': crisis,
        'Strategy': 'Breakout Strategy (DIA)',
        'Max Drawdown (%)': strat2_metrics[0],
        'Recovery Time (Days)': strat2_metrics[1],
        'Total Return (%)': strat2_metrics[2],
    })
    metrics_data.append({
        'Crisis Period': crisis,
        'Strategy': 'Breakout Strategy (DJIA Constituents)',
        'Max Drawdown (%)': strat3_metrics[0],
        'Recovery Time (Days)': strat3_metrics[1],
        'Total Return (%)': strat3_metrics[2],
    })

# Create the metrics DataFrame
metrics_df = pd.DataFrame(metrics_data)
```

```python
# Function to compute metrics for a given period
def get_metrics(df, start, end, value_col, drawdown_col):
    # Filter data for the crisis period
    data = df.loc[start:end]
    if data.empty:
        return None, None, None

    # Maximum Drawdown: Minimum drawdown value (already in percentage)
    max_drawdown = data[drawdown_col].min() * 100  # Convert to percentage
    trough_date = data[drawdown_col].idxmin()

    # Recovery Time: Days to recover to pre-crisis peak (or "Not Recovered" if not recovered)
    peak_date = data[value_col][:trough_date].idxmax()
    peak_value = data[value_col][peak_date]
    recovery_data = df[value_col][trough_date:]
    recovery_date = recovery_data[recovery_data >= peak_value].index
    recovery_days = (recovery_date[0] - trough_date).days if not recovery_date.empty else "Not Recovered"

    return max_drawdown, recovery_days, trough_date

# Compute metrics for each crisis period and strategy
metrics = []
for crisis, info in crisis_periods.items():
    start, end = info["dates"]
    start = pd.to_datetime(start)
    end = pd.to_datetime(end)

    # Strategy 1
    s1_drawdown, s1_recovery, s1_trough = get_metrics(Strat1, start, end, 'Cumulative_Return', 'Drawdown')
    s2_drawdown, s2_recovery, s2_trough = get_metrics(Strat2, start, end, 'Cumulative_Return', 'Drawdown')
    s3_drawdown, s3_recovery, s3_trough = get_metrics(Strat3, start, end, 'Cumulative_Return', 'Drawdown')

    metrics.append({
        'crisis': crisis,
        's1_drawdown': round(s1_drawdown, 2) if s1_drawdown is not None else None,
        's1_recovery': s1_recovery,
        's1_trough': s1_trough,
        's2_drawdown': round(s2_drawdown, 2) if s2_drawdown is not None else None,
        's2_recovery': s2_recovery,
        's2_trough': s2_trough,
        's3_drawdown': round(s3_drawdown, 2) if s3_drawdown is not None else None,
        's3_recovery': s3_recovery,
        's3_trough': s3_trough
    })

# Plot the combined drawdown graph
plt.figure(figsize=(16, 10))  # Match the size of other graphs
plt.fill_between(Strat1.index, Strat1['Drawdown'], 0, color='green', alpha=0.1)
plt.plot(Strat1.index, Strat1['Drawdown'], label='Buy & Hold (DIA)', color='green', linewidth=2)
plt.fill_between(Strat2.index, Strat2['Drawdown'], 0, color='blue', alpha=0.1)
plt.plot(Strat2.index, Strat2['Drawdown'], label='Breakout Strategy (DIA)', color='blue', linewidth=2)
plt.fill_between(Strat3.index, Strat3['Drawdown'], 0, color='orange', alpha=0.1)
plt.plot(Strat3.index, Strat3['Drawdown'], label='Breakout Strategy (DJIA Constituents)', color='orange', linewidth=2)

# Shade the crisis periods
for crisis, info in crisis_periods.items():
    start, end = info["dates"]
    plt.axvspan(start, end, color=info["color"], alpha=0.3, label=crisis)

# Add annotations for max drawdown and recovery time
for m in metrics:
    if m['s1_trough'] is not None:
        s1_y = Strat1.loc[m['s1_trough'], 'Drawdown']
        plt.text(m['s1_trough'], s1_y - 0.05, f'MD: {m["s1_drawdown"]}%\nRT: {m["s1_recovery"]} days', 
                 color='green', fontsize=12, ha='center', va='top')
    if m['s2_trough'] is not None:
        s2_y = Strat2.loc[m['s2_trough'], 'Drawdown']
        plt.text(m['s2_trough'], s2_y + 0.05, f'MD: {m["s2_drawdown"]}%\nRT: {m["s2_recovery"]} days', 
                 color='blue', fontsize=12, ha='center', va='bottom')
    if m['s3_trough'] is not None:
        s3_y = Strat3.loc[m['s3_trough'], 'Drawdown']
        plt.text(m['s3_trough'], s3_y + 0.05, f'MD: {m["s3_drawdown"]}%\nRT: {m["s3_recovery"]} days', 
                 color='black', fontsize=12, ha='center', va='bottom')  # Black for visibility against yellow

# Customize the plot
plt.title("[Figure 7] Drawdown Comparison: Buy & Hold vs Breakout Strategies", fontsize=22)
plt.xlabel("Date", fontsize=18)
plt.ylabel("Drawdown", fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(-0.6, 0)
plt.legend(
    loc='upper left', 
    fontsize=12,
    borderaxespad=0,
    frameon=True,
    edgecolor='black',
    framealpha=1
)
plt.grid(True, linestyle='--', alpha=0.7)

# Add figure caption below the plot
fig_caption = "Figure: This line plot compares the drawdown over time for Buy & Hold DIA, Breakout DIA, and Breakout DJIA Constituents strategies, with max drawdown (MD) and recovery time (RT) annotated during liquidity shocks."
plt.figtext(0.5, 0.05, fig_caption, wrap=True, horizontalalignment='center', fontsize=14)

# Adjust layout to prevent clipping of the caption
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)

plt.show()
```
